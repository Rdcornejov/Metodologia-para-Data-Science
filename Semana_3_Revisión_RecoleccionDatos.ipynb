```python
 # Importar librerías necesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Cargar el dataset

data = pd.read_csv("https://github.com/Rdcornejov/Metodologia-para-Data-Science/raw/main/Dataset/sales_data_sample.csv",encoding='ISO-8859-1')
```

```python
data.head(10)
```

```python
data.info()
```

```python
# Identificar valores faltantes
print("Valores faltantes en cada columna:")
print(data.isna().sum())
```

```python
# Verificar problemas de formato en la columna 'Fecha Registro'
data['ORDERDATE'] = pd.to_datetime(data['ORDERDATE'])
data.info()
```

```python
# Verificar consistencia en la columna 'STATUS'
print("\nValores únicos en la columna 'STATUS':")
print(data['STATUS'].unique())
```

```python
# Visualizar distribución de precios para detectar anomalías
plt.figure(figsize=(10, 6))
sns.histplot(data['PRICEEACH'].dropna(), bins=10, kde=True)
plt.title('Distribución de Precios')
plt.xlabel('PRICEEACH')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Identificar posibles errores en la columna 'STATE'
print("\nValores nulos en la columna 'STATE':")
print(data['STATE'].isna().sum())
```

```python
# Identificar posibles errores en la columna 'STATE'
print("\nValores nulos en la columna 'TERRITORY':")
print(data['TERRITORY'].isna().sum())
```

```python
# Verificar consistencia en la columna 'TERRITORY'
print("\nValores únicos en la columna 'TERRITORY':")
print(data['TERRITORY'].unique())
```
